```python
class BankAccount:

    # class constructor
    def __init__(self, username: str, password: str, balance: float = 0) -> None:
        self.username: str = username.strip()
        self.password: str = password.strip()
        self.balance: float = balance

    # non static method to deposit
    def deposit(self, amount: float) -> None:
        if amount > 0:
            self.balance += amount
        else:
            print("Invalid amount is entered")

    # non static method to withdraw
    def withdraw(self, amount: float) -> None:
        if amount > 0 and amount <= self.balance:
            self.balance -= amount
        else:
            print("Invalid amount is entered")

    # non static method to print the info
    def print_Account_details(self) -> None:
        print("The username -->", self.username)
        print("The password -->", "دي اسرار ميصحش")
        print("The balance -->", self.balance)
```

```python
class Bank:

    # class constructor
    def __init__(self) -> None:
        self.bank_accounts: list[BankAccount] = []

    # non static method to add customer
    def add_bank_account(self, ba: BankAccount) -> None:
        self.bank_accounts.append(ba)
        print("The bank Account is added successfully")

    # non static method to remove customer
    def remove_bank_account(self, ba: BankAccount) -> None:
        if ba in self.bank_accounts:
            self.bank_accounts.remove(ba)
            print("The bank Account is removed successfully")
        else:
            print("There is no Bank Account to remove")

    # Search account by the username
    def search_accounts(self, username: str) -> BankAccount | None:
        for account in self.bank_accounts:
            if account.username == username:
                return account
        return None

    # using the dunder methods, return number of accounts
    def __len__(self):
        return len(self.bank_accounts)
```

```python
# Create accounts
acc1 = BankAccount(" ahmed ", " 1234 ", 1000)
acc2 = BankAccount("sara", "abcd", 500)

print("\n--- Deposit Test ---")
acc1.deposit(500)
print(acc1.balance == 1500)
acc1.deposit(-100)
print(acc1.balance == 1500)

print("\n--- Withdraw Test ---")
acc1.withdraw(300)
print(acc1.balance == 1200)
acc1.withdraw(1200)
print(acc1.balance == 0)
acc1.withdraw(100)
print(acc1.balance == 0)

print("\n--- Print Account Details ---")
acc1.print_Account_details()


bank = Bank([])
print("\n--- Add Accounts ---")
bank.add_bank_account(acc1)
bank.add_bank_account(acc2)
print(len(bank.bank_accounts) == 2)

print("\n--- Search Accounts ---")
found = bank.search_accounts("sara")
print(found == acc2)
not_found = bank.search_accounts("mohamed")
print(not_found is None)

print("\n--- Remove Accounts ---")
bank.remove_bank_account(acc1)
print(len(bank.bank_accounts) == 1)
bank.remove_bank_account(acc1)
```
