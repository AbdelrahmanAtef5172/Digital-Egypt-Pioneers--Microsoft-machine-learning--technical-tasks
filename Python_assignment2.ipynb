```python
#Some Notes
# THE STATIC VARIABLES CAN BE ACCESSED THROUGH THE NAME OF THE OBJECT OR THE CLASS
# THE STATIC METHODS CAN ONLY BE ACCESSED THROUGH THE NAME OF THE CLASS AND IF THEY ARE CALLED THROUGH OBJECT
# (SELF) WILL BE PASSED AS AN ARGUMENT AND ERROR WILL APPEAR AS THE STATIC METHOD DOESNT WAIT FOR ANY ARGUMENTS
# DUNDER METHODS LIKE __str__ , __eq__ , __len__  to customize their behaviour according to your Class
# __str__ → readable string representation.
# __eq__ → defines equality comparison.
# __len__ → defines behavior for len()
```

```python
#Example 1 during session
class Car :
  # Static Variables
  number_cars_created = 0

  # class constructor
  def __init__ (self , model: int , brand:str , color:str):
    self.model = model
    self.brand = brand.strip()
    self.color = color.strip()
    Car.number_cars_created +=1

 # method to change car color
  def change_color(self , color:str):
     self.color = color.strip()
 # method to change the car model
  def change_model(self , model:int):
     self.model = model
 # method to change the car brand
  def change_brand(self,brand:str):
     self.brand = brand.strip()
  # static method to total number of cars
  def get_number_cars() :
    return Car.number_cars_created

  #getter and setters
  def get_color(self) : return self.color
  def get_brand(self) : return self.brand
  def get_model(self) : return self.model

  # method to print the car details
  def print_info(self):
    print("The car model is " , self.model)
    print("The car Brand is " , self.brand)
    print("The car color is " , self.color)
    print("-"*40)

 # overriding the dunder methods
  def __str__(self):
    return f"model-->{self.model} , brand-->{self.brand}, color-->{self.color}"
  def __len__(self):
    return len(self.brand + self.color)
  def __eq__(self,other) -> bool:
    return self.model == other.model and self.brand == other.brand and self.color == other.color


# testing the class
myCar = Car(2025 , "BMW_M4","Black")
myCar.print_info()
myCar.change_brand("BMW M3")
myCar.change_color("Grey")
myCar.change_model(2024)
myCar.print_info()
print("The number of cars Created -->", Car.get_number_cars())

myFriendCar = Car(2026 , "Mercedes","White")
print("The number of the cars created -->" , Car.get_number_cars())

# testing the dunder methods
print(myCar)
print("-"*30)
print(len(myCar)>len(myFriendCar))
# will compare the length of the strings
#(BMW M4Black)(MercedesWhite) so it so it should return False

#car comparisons
car1 = Car(2010 , "Skoda","Black")
car2 = Car(2010 , "Skoda","Black")
if(car1 == car2):
  print("The two cars are the same ")
else :
  print("The two cars are diff")

```

```python
from typing import List

#Example 2 assignemnet
class Book:
    def __init__(self, author: str, title: str) -> None:
        self.author: str= author.strip()
        self.title: str = title.strip()

    def __str__(self) -> str:
        return self.title


class Library:
    def __init__(self, books: List[Book], wifi: bool = True) -> None:
        self.books: List[Book] = books
        self.wifi: bool = wifi

    def add_book(self, book: Book) -> None:
        self.books.append(book)
        print("The book is added successfully")

    def add_books(self, books: List[Book]) -> None:
        for book in books:
            self.books.append(book)

    def display_titles(self) -> None:
        for book in self.books:
            print(book.title)

    def display_books(self) -> None:
       if self.books:
          for book in self.books:
            print(book.title, book.author)
       else:
         print("There are no books availble in the Library")


    # implementing it in two ways
    def get_number_of_books(self):
       return len(self.books)
    def __len__(self):
         return len(self.books)


    def search_book(self, title: str) -> Book:
      for book in self.books:
        if book.title == title.strip():
            return book
      print("Book is not found")
      return None


    def remove_book(self, title: str) -> None:
      for book in self.books:
        if book.title == title.strip():
            self.books.remove(book)
            print("The book is removed successfully")
            return
      print("The book doesnt exist to remove")

```

```python
book1 = Book(author="George Orwell", title="1984")
book2 = Book(author="Harper Lee", title="To Kill a Mockingbird")
book3 = Book(author="J.K. Rowling", title="Harry Potter")
library = Library(books=[])
library.add_book(book1)
library.add_books([book2, book3])
library.display_titles()

###########
print("-"*40)
library.display_books()
print("-"*40)
print("Number of books:", library.get_number_of_books())
print("Length of library using len():", len(library))
# checking for book exists
found_book = library.search_book("1984")
if found_book:
    print("Found book:", found_book.title, "-", found_book.author)
#checking for book doesnt exist
library.search_book("The Alchemist")
print("-"*40)
# removing book exists
library.remove_book("1984")
library.display_books()
print("-"*40)
#removing book doesnt exist
library.remove_book("The Alchemist")
```
