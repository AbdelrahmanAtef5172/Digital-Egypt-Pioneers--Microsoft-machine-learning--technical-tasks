```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df  = pd.read_csv("diamonds.csv")
```

```python
df.head(10)
```

```python
df.info()
```

```python
df = df.drop('Unnamed: 0', axis=1)
```

```python
numerical_features = ['carat', 'depth', 'table', 'price', 'x', 'y', 'z']
for feature in numerical_features:
    plt.figure(figsize=(8, 6))
    sns.histplot(df[feature], kde=True)
    plt.title(f'Distribution of {feature}')
    plt.xlabel(feature)
    plt.ylabel('Frequency')
    plt.show()
```

```python
categorical_features = ['cut', 'color', 'clarity']
for feature in categorical_features:
    plt.figure(figsize=(10, 6))
    sns.countplot(x=feature, data=df)
    plt.title(f'Count Plot of {feature}')
    plt.xlabel(feature)
    plt.ylabel('Count')
    plt.show()
```

```python
print(df['clarity'].value_counts())
# we may drop the last 3 categories as they are so rare
```

```python
plt.figure(figsize=(10, 7))
sns.scatterplot(x='carat', y='price', data=df, alpha=0.7)
plt.title('Carat vs. Price')
plt.xlabel('Carat')
plt.ylabel('Price')
plt.grid(True)
plt.show()
# exponential raise
```

```python
for feature in categorical_features:
    plt.figure(figsize=(10, 6))
    sns.boxplot(x=feature, y='price', data=df)
    plt.title(f'Price Distribution by {feature}')
    plt.xlabel(feature)
    plt.ylabel('Price')
    plt.show()
```

```python
numerical_df = df.select_dtypes(include=[np.number])
correlation_matrix = numerical_df.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=.5)
plt.title('Correlation Matrix of Numerical Features')
plt.show()
# There are many columns to drop due to high colleration
```
